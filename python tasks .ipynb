# Calculate the top 10 stores with the highest average daily sales.

```python
# Merge orders and order_items to get sales data per order
sales_per_order = pd.merge(orders, order_items, on='order_id')
sales_per_order
```

```python
# Convert 'order_purchase_timestamp' to datetime and extract the date
sales_per_order['order_purchase_timestamp'] = pd.to_datetime(sales_per_order['order_purchase_timestamp'])
sales_per_order['order_date'] = sales_per_order['order_purchase_timestamp'].dt.date
sales_per_order
```

```python
# Aggregate sales per day per seller
daily_sales = sales_per_order.groupby(['seller_id', 'order_date']).agg(daily_sales=('price', 'sum')).reset_index()
daily_sales
```

```python
# Calculate average daily sales per seller
average_daily_sales = daily_sales.groupby('seller_id').agg(average_sales=('daily_sales', 'mean')).reset_index()
average_daily_sales
```

```python
# Sort and get the top 10 sellers with the highest average daily sales
top_sellers_average_daily_sales = average_daily_sales.sort_values(by='average_sales', ascending=False).head(10)
top_sellers_average_daily_sales
```

```python
# Plotting the top 10 sellers based on average daily sales

plt.figure(figsize=(10,7))
ax= sns.barplot(data=top_sellers_average_daily_sales, x='average_sales', y='seller_id', palette='crest')
plt.xlabel('Average Daily Sales ($)')
plt.ylabel('Seller ID')
ax.bar_label(ax.containers[0]) 
plt.title('Top 10 Sellers by Average Daily Sales')
plt.show()
```

# Calculate the percentage of monthly growth in sales for each store.

```python
# Aggregate sales per month per seller
monthly_sales_sellers = sales_per_order.groupby(['seller_id', sales_per_order['order_purchase_timestamp'].dt.to_period('M')]).agg(
    monthly_sales=('price', 'sum')).reset_index()
monthly_sales_sellers
```

```python

# Calculate month-over-month percentage change in sales for each seller
monthly_sales_sellers['monthly_growth'] = monthly_sales_sellers.groupby('seller_id')['monthly_sales'].pct_change() * 100
monthly_sales_sellers
```

```python

# Pivot table for clearer view by seller_id and month
monthly_growth_pivot = monthly_sales_sellers.pivot(index='order_purchase_timestamp', columns='seller_id', values='monthly_growth')

monthly_growth_pivot.tail(10)  # Display the last few entries to show recent data trends
```

```python
# Select a few sellers that have relatively complete data for visualization
selected_sellers = monthly_sales_sellers['seller_id'].value_counts().nlargest(3).index.tolist()
selected_sellers
```

```python

# Filter the monthly growth data for these selected sellers
selected_monthly_growth = monthly_sales_sellers[monthly_sales_sellers['seller_id'].isin(selected_sellers)]
selected_monthly_growth
```

```python

# Pivot table for clearer view by seller_id and month
selected_growth_pivot = selected_monthly_growth.pivot(index='order_purchase_timestamp', columns='seller_id', values='monthly_growth')
selected_growth_pivot
```

```python

# Plotting the monthly growth trend for selected sellers
plt.figure(figsize=(14, 8))

for seller in selected_sellers:
    sns.lineplot(x=selected_growth_pivot.index.astype(str), 
             y=selected_growth_pivot[seller].to_numpy(), 
             marker='o', 
             label=f'Seller {seller}')

    
plt.title('Monthly Growth Trends for Selected Sellers')
plt.xlabel('Month-Year')
plt.ylabel('Monthly Growth (%)')
plt.legend(title='Seller ID')
plt.xticks(rotation=90)
#plt.grid(True)
plt.tight_layout()
plt.show()
```

# Conduct a cohort analysis on customers’ orders based on the month of their first purchase and analyze their behavior over time

```python
# Convert order purchase timestamp to datetime 
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
```

```python
# Create a dataframe to identify each customer's first purchase
first_purchase = orders.groupby('customer_id').order_purchase_timestamp.min().reset_index()
first_purchase.columns = ['customer_id', 'first_purchase_date']
first_purchase['first_purchase_date'] = first_purchase['first_purchase_date'].dt.to_period('M')
first_purchase
```

```python
# Merge this data back to the orders dataset to assign each order to a cohort
orders_cohort = pd.merge(orders, first_purchase, on='customer_id', how='left')
orders_cohort
```

```python
# Assign Order Period as the difference in months from first purchase
orders_cohort['order_period'] = (orders_cohort.order_purchase_timestamp.dt.to_period('M') - orders_cohort.first_purchase_date).apply(lambda x: x.n)
orders_cohort
```

```python

# Aggregate cohort data by first purchase date and order period
cohort_data = orders_cohort.groupby(['first_purchase_date', 'order_period']).agg(n_customers=('customer_id', 'nunique')).reset_index()
cohort_data
```

```python

# Pivot to create the cohort table
cohort_pivot = cohort_data.pivot_table(index='first_purchase_date', columns='order_period', values='n_customers')

# Fill NaN with 0 for visualization
cohort_pivot.fillna(0, inplace=True)

cohort_pivot
```

```python
# Plotting the heatmap
plt.figure(figsize=(11, 7))
sns.heatmap(cohort_pivot, annot=True, cmap='crest', fmt='g')
plt.title('Cohort Analysis - Customer Retention Over Time')
plt.xlabel('Months Since First Purchase')
plt.ylabel('Cohort (First Purchase Month)')
plt.show()
```

```python
import matplotlib.colors as mcolors

# Heatmap using matplotlib for the initial cohort sizes
fig, ax = plt.subplots(figsize=(12, 10))
cmap = mcolors.LinearSegmentedColormap.from_list("", ["green", "blue"])
cax = ax.matshow(cohort_pivot, interpolation='nearest', cmap=cmap)
fig.colorbar(cax)
ax.set_xticklabels(['']+cohort_pivot.columns.tolist(), rotation=90)
ax.set_yticklabels(['']+cohort_pivot.index.astype(str).tolist())
ax.set_xlabel('Months since First Purchase')
ax.set_ylabel('Cohort (First Purchase Month)')
ax.set_title('Cohort Analysis - Customers in Initial Purchase Month')
plt.show()
```

```python
# Create a dataframe to identify each customer's first purchase year
first_purchase_year = orders.groupby('customer_id').order_purchase_timestamp.min().dt.year.reset_index()
first_purchase_year.columns = ['customer_id', 'first_purchase_year']
first_purchase_year
```

```python

# Merge this data back to the orders dataset to assign each order to a cohort based on the first purchase year
orders_cohort_year = pd.merge(orders, first_purchase_year, on='customer_id', how='left')
```

```python

# Assign Order Year for easier grouping by year
orders_cohort_year['order_year'] = orders_cohort_year['order_purchase_timestamp'].dt.year
```

```python

# Create a year period column representing the number of years since the first purchase
orders_cohort_year['year_period'] = orders_cohort_year['order_year'] - orders_cohort_year['first_purchase_year']
```

```python

# Aggregate cohort data by first purchase year and year period
cohort_data_yearly = orders_cohort_year.groupby(['first_purchase_year', 'year_period']).agg(n_customers=('customer_id', 'nunique')).reset_index()
```

```python
# Pivot to create the cohort table
cohort_pivot_yearly = cohort_data_yearly.pivot_table(index='first_purchase_year', columns='year_period', values='n_customers')
```

```python

# Normalize the cohort table to show retention rates
cohort_pivot_yearly_normalized = cohort_pivot_yearly.divide(cohort_pivot_yearly[0], axis=0)
```

```python

cohort_pivot_yearly_normalized.fillna(0, inplace=True)  # Fill NaN with 0 for visualization
cohort_pivot_yearly_normalized
```

##### It appears that our analysis only shows data for the initial purchase year across the cohorts from 2016 to 2018, with no subsequent year data available for any cohorts. This indicates that either the customers did not make any purchases in the years following their first purchase, or there is no data available in the dataset to track these subsequent purchases.

##### Given this limitation, we're unable to perform a meaningful analysis of yearly customer retention beyond the initial purchase year. This suggests that either the dataset is truncated or customer retention beyond the first year is not captured

```python
# Convert 'order_purchase_timestamp' to datetime and extract the week and year
orders['order_week'] = orders['order_purchase_timestamp'].dt.to_period('W')

# Aggregate orders per week
weekly_orders = orders.groupby('order_week').size().reset_index(name='number_of_orders')

# Plotting the weekly customer order trends
plt.figure(figsize=(14, 7))
plt.plot(weekly_orders['order_week'].astype(str), weekly_orders['number_of_orders'], marker='o', linestyle='-')
plt.title('Weekly Customer Order Trends')
plt.xlabel('Week of Year')
plt.ylabel('Number of Orders')
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()
plt.show()
```
